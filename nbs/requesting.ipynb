```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def bytes2array(data):
    data = data.decode('utf8')
    data = base64.b64decode(data)
    array = np.fromstring(data, np.uint8)
    image = cv2.imdecode(array, cv2.IMREAD_UNCHANGED)
    return image 
 

```
